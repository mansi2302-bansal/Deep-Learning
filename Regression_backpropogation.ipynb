## Implementing MLP hard code


1. here we are initializing the number of parameters required.
2. if we have 2 input features and 3 nodes in hidden layer 1, then number parameters required in layer 1 = 2x3 = 6.
3. We have just formed arrays/ vectors of dimensions (l-1,l) for weights and (l,1) for bias.
4. Now, next step is to calculate the Weighted Sum.

```python
def initialize_param(layer_dims):
  np.random.seed(3)
  parameters = {}

  for l in range(1,len(layer_dims)):
    parameters["W"+str(l)] = np.ones((layer_dims[l-1], layer_dims[l]))*0.1
    parameters["b"+str(l)] = np.ones((layer_dims[l], 1))

  return parameters
```

```python
## check
initialize_param([2,2,1])
```

1. From the output we can see that, the NN has 2 nodes in input layer, 2 nodes in hidden layer and 1 in output layer.
2. W1 has 2x2 shape => 4 weights.
3. 2 biases in layer 1
4. W2 has (2,1) => 2 weights
5. 1 bias.


```python
# Now we have to find the weighted sum
def linear_forward(A_prev, w, b):  # A_prev matrix is the output we get after a layer. eg A0 is the output we got from layer 1
   Z = np.dot(w.T, A_prev) + b

   return Z
```

```python
# Forward propogation
def L_forward(X, parameters):
  A = X
  num_layers = len(parameters)//2

  for i in range(1,num_layers+1):
    A_prev = A
    wi = parameters["W"+str(i)]
    bi = parameters["b"+str(i)]
    #print(f"A{i-1} =: ",A_prev)
    #print(f"w{i} =: ",wi)
    #print(f"b{i} =: ",bi)

    A = linear_forward(A_prev, wi,bi)
    #print("A"+str(i)+" =: ", A)


  return A, A_prev
```

1. Now, We have to find weighted sum and W1 has 2x2 shape for which X has to be something in form : (2,) for multiplication to hold.
2. X =: (2,1)
3. Z1 = wTX + b
4. A =: [z1,z2]T

```python
X = df[['cgpa', 'profile_score']].values[0].reshape(2,1) # Shape(no of features, no. of training example)
y = df[['lpa']].values[0][0]

# Parameter initialization
parameters = initialize_param([2,2,1])

y_hat,A1 = L_forward(X, parameters)
```

1. Now, we have to update the parameters using gradient descent algorithm

```python
def update_parameters(parameters,y,y_hat,A1,X):
  parameters['W2'][0][0] = parameters['W2'][0][0] + (0.001 * 2 * (y - y_hat)*A1[0][0])
  parameters['W2'][1][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat)*A1[1][0])
  parameters['b2'][0][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat))

  parameters['W1'][0][0] = parameters['W1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[0][0])
  parameters['W1'][0][1] = parameters['W1'][0][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[1][0])
  parameters['b1'][0][0] = parameters['b1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0])

  parameters['W1'][1][0] = parameters['W1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[0][0])
  parameters['W1'][1][1] = parameters['W1'][1][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[1][0])
  parameters['b1'][1][0] = parameters['b1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0])
```

Now, Here we have formed 1 code that is iterating over each row and calculating loss after each computation

```python
# epochs implementation

parameters = initialize_param([2,2,1])
epochs = 5

for i in range(epochs):

  Loss = []

  for j in range(df.shape[0]):

    X = df[['cgpa', 'profile_score']].values[j].reshape(2,1) # Shape(no of features, no. of training example)
    y = df[['lpa']].values[j][0]

    # Parameter initialization


    y_hat,A1 = L_forward(X,parameters)
    y_hat = y_hat[0][0]

    update_parameters(parameters,y,y_hat,A1,X)

    Loss.append((y-y_hat)**2)

  print('Epoch - ',i+1,'Loss - ',np.array(Loss).mean())

```
